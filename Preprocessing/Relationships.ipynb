# Input =  {r1, r2, b1, a2, RatioTotalArea, frac}
Here an attempt is made to understand how these inputs change eta c. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
sns.set()
```

```python
def import_data(name='HybridEllipsePercolation.txt', sep1=" ", header1=None, shuffle=True):
    data = pd.read_csv(name, sep=sep1, header=header1)
    data.columns = ["r1", "2a2", "r2", "frac", "Nc", "Nc Std. Dev", "eta c" ]
    # data.reset_index(inplace=True)
    
    if shuffle:
        data = data.sample(frac=1).reset_index(drop=True)
    return data

def single_input_vs_output(dataset, input_column, output_column="eta c", plot=False, output=False):
    reduced_dataset = dataset.drop_duplicates(subset=input_column)
    # data_range = reduced_dataset[input_column].to_numpy()
    if output:
        print("{col} range is: {rng}    (output is {out})".format(col=input_column, rng=len(reduced_dataset), out=output_column))

    if plot:
        ax1 = reduced_dataset.plot.scatter( x=input_column,
                        y=output_column,
                        c='DarkBlue')
        return ax1
        # return sns.scatterplot(data=reduced_dataset, x=input_column, y=output_column)
    
    return reduced_dataset[[input_column, output_column]]

def split_data(dataset):
    train_dataset = dataset.sample(frac=0.6, random_state=0)
    valid_and_test_dataset = dataset.drop(train_dataset.index)
    test_dataset = valid_and_test_dataset.sample(frac=0.5, random_state=0)
    validation_dataset = valid_and_test_dataset.drop(test_dataset.index)
    return train_dataset, test_dataset, validation_dataset

def split_features_labels(data, label_column='eta c'):
    features = data
    labels = data.pop(label_column)
    return features, labels
```

```python
rawdata = import_data(shuffle=False)
dataset = rawdata.copy()

# remove irrelevant columns
# dataset.pop("Nc Std. Dev")

# check for missing values
dataset.isna().sum()
# drop missing values
dataset = dataset.dropna()

dataset["a1"]=1/2
dataset["b1"]=dataset["a1"]/dataset["r1"]
dataset["a2"]=dataset["2a2"]/2
dataset["b2"]=dataset["a2"]/dataset["r2"]
dataset["area1"]=dataset["a1"]*dataset["b1"]*np.pi
dataset["area2"]=dataset["a2"]*dataset["b2"]*np.pi
dataset["TotalArea"] = dataset["area1"] + dataset["area2"]
dataset["RatioTotalArea"] = dataset["area1"]*(1-dataset["frac"]) + dataset["area2"]*(dataset["frac"])
dataset.tail()
```

```python
r1_range = single_input_vs_output(dataset, "r1")
r2_range = single_input_vs_output(dataset, "r2")
frac_range = single_input_vs_output(dataset, "frac")
a1_range = single_input_vs_output(dataset, "a1")
b1_range = single_input_vs_output(dataset, "b1")
a2_range = single_input_vs_output(dataset, "a2")
b2_range = single_input_vs_output(dataset, "b2")
_2a2_range = single_input_vs_output(dataset, "2a2")
```

## Splitting Eta C

```python
def split_data2(dataset, seed, train_ratio=0.6, shuffle=True, highest_at_top=False, no_outliers=False, center_split=0.6, abv_06_test_ratio=0.8):
    df = dataset.copy()
    
    tot_num_of_samples = int(len(df)*train_ratio)
    # split the data
    df_abv = df[df["eta c"] > center_split]
    df_bel = df[df["eta c"] <= center_split]
    abv_06 = len(df_abv)
    bel_06 = len(df_bel)
    print("Maximum training samples above EtaC=0.6 is", int(abv_06*abv_06_test_ratio))
    print("Number of samples required for EtaC<=0.6 or >0.6 is", tot_num_of_samples/2)
    # num of samples above and below 0.6 must be equal
    # thus, there is a maximum number of samples available 
    # for training even when not using the entire data set
    if tot_num_of_samples/2 <= abv_06*abv_06_test_ratio:
        # abv_06_samples = tot_num_of_samples/2
        # bel_06_samples = tot_num_of_samples/2
        print("EtaC>0.6 does not limit data.")
        abv_ratio = int((tot_num_of_samples/2)/abv_06)
        bel_ratio = int((tot_num_of_samples/2)/bel_06)
    else:
        # abv_06_samples = abv_06*abv_06_test_ratio
        # bel_06_samples = abv_06*abv_06_test_ratio
        abv_ratio = (abv_06*abv_06_test_ratio)/abv_06
        bel_ratio = (abv_06*abv_06_test_ratio)/bel_06
        print("EtaC>0.6 does limit data to {} samples each (above or below 0.6).". format(int(abv_06*abv_06_test_ratio)))

    # need to split the number of samples into above 0.6 and below.
    # Maybe will have a maximum number of samples. 
    # does not definitely include last few
    
    # shuffle the datasets
    train_dataset = df_abv.sample(frac=abv_ratio, random_state=seed).append(df_bel.sample(frac=bel_ratio, random_state=seed))
    test_dataset = df.drop(train_dataset.index)

    print()
    print("Number of Training set samples: ", len(train_dataset))
    print("Above 0.6: {x1}    |   Below or equal to 0.6: {x2}".format(x1=len(train_dataset[train_dataset["eta c"] > center_split]), x2=len(train_dataset[train_dataset["eta c"] <= center_split])))
    print("Number of Testing set samples: ", len(test_dataset))
    print("Above 0.6: {x1}    |   Below or equal to 0.6: {x2}".format(x1=len(test_dataset[test_dataset["eta c"] > center_split]), x2=len(test_dataset[test_dataset["eta c"] <= center_split])))

    return train_dataset, test_dataset
```
